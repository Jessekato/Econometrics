```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulation settings
np.random.seed(42)
n = 100
epsilon = np.random.normal(0, 1, n)

# Case 1: Unit root (rho = 1)
y_unit_root = [0]
for t in range(1, n):
    y_unit_root.append(y_unit_root[t-1] + epsilon[t])
y_unit_root = pd.Series(y_unit_root, name="Unit Root ( = 1)")

# Case 2: Explosive root (rho = 1.5)
y_explosive = [0]
for t in range(1, n):
    y_explosive.append(1.5 * y_explosive[t-1] + epsilon[t])
y_explosive = pd.Series(y_explosive, name="Explosive Root ( = 1.5)")

# Plot both series
plt.figure(figsize=(12, 6))
plt.plot(y_unit_root, label='Unit Root ( = 1)')
plt.plot(y_explosive, label='Explosive Root ( = 1.5)', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5, linestyle=':')
plt.title("Comparison of Unit Root vs Explosive Root Time Series")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
